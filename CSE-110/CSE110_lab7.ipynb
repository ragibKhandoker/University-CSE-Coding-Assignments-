```python
# Task 1
def bubbleSortAscending(lst):
    for item in range(len(lst)):
        for value in range(len(lst)-1-item):
            if lst[value+1] < lst[value]:
                lst[value] , lst[value+1] = lst[value+1] , lst[value]
    return lst
my_list = [10,1,20,3,6,2,5,11,15,2,12,14,17,18,29]
print(bubbleSortAscending(my_list))
```

```python
# Task 2
def selectionSortAscending(lst):
    for curr_i in range(len(lst)):
        min_i = curr_i
        for value in range(curr_i+1, len(lst)):
            if lst[value] < lst[min_i]:
                min_i = value
        lst[curr_i] , lst[min_i] = lst[min_i], lst[curr_i]
    return lst
my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
print(selectionSortAscending(my_list))
```

```python
# Task 7
list1= []
list2= []
element1=int(input('Element no for 1st list: '))
element2=int(input('Element no for 2nd list: '))
for item1 in range(0,element1):
    item1=int(input('Enter element for first list: '))
    list1.append(item1)
for item2 in range(0,element2):
    item2=int(input('Enter element for second list: '))
    list2.append(item2)
new_list=list1+list2

def bubbleSortAscending(new_list):
    for item in range(len(new_list)):
        for value in range(len(new_list)-1-item):
            if new_list[value+1] < new_list[value]:
                new_list[value] , new_list[value+1] = new_list[value+1] , new_list[value]
    return new_list
sorted_newlist = bubbleSortAscending(new_list)

print('Sorted list=',sorted_newlist)
length=len(sorted_newlist)

if length%2!=0:
    num=int((length+1)/2)

    print('Median=',sorted_newlist[num-1])
else:
    index= int((length)//2)-1
    index1=int(((length)/2)+1)
    median=float((sorted_newlist[index]+sorted_newlist[index1])/2)
    print('Median=',median)
```

```python
# Task 8
def function_name(funcn) :
    var1=0
    var2=1
    min_sum = abs(funcn[0] + funcn[1])
    for item1 in range(0,len(funcn)-1) :
        for item2 in range(item1+1,len(funcn)) :
            sum1 = abs(funcn[item1] + funcn[item2])
            if min_sum > sum1 :
                min_sum = sum1
                var1 = item1
                var2 = item2
    print("Two pairs which have the smallest sum =", funcn[var1], "and", funcn[var2])
funcn=[]
element=int(input('Element no for the list: '))
for item1 in range(0,element):
    item1=int(input('Enter element for first list: '))
    funcn.append(item1)
function_name(funcn)
```

```python
# Task 9
import math
def function_name(funcnx,funcny) :
    return math.sqrt(funcnx**2 + funcny**2)
points = [(1,7), (4,5), (-1,7), (-2,0), (1,1), (5,-1)]
var1 = len(points)
dist = []
for item in range(var1) :
    dist.append(function_name(points[item][0], points[item][1]))
min_dist = min(dist)
min_idx = dist.index(min_dist)
print("Minimum distance =", min_dist)
print("Here the closest point is " + str(points[min_idx]) + " which has a distance of " + str(min_dist) + " from the origin.")
```
