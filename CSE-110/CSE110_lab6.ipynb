```python
# Task 4
def function_name(funcn):
    upper = 0
    lower = 0
    for symbol in funcn:
        if symbol.isupper():
            upper += 1
        if symbol.islower():
            lower += 1
    print("No. of Uppercase characters: ", upper)
    print("No. of Lowercase Characters: ", lower)
function_name(input('Enter a string :'))
```

```python
# Task 5
def calculate_tax(age, salary, current_job):
    if age < 18:
        return 0
    if current_job == 'PRESIDENT OF THE COMPANY':
        return 0
    if salary < 10000:
        return 0
    if 10000 <= salary <= 20000:
        return salary * 0.05
    else:
        return salary * 0.1
year = int(input('Enter your age :'))
money = int(input('Enter your salary :'))
job = input('Enter your job name :')
calculate_tax(year, money, job)
```

```python
# Task 7
def show_palindrome(number):
    answer = ''
    for item in range(1, number * 2):
        if item <= number:
            answer += str(item)
        else:
            answer += str(number * 2 - item)
    print(answer)
show_palindrome(int(input('Enter a number :')))
```

```python
# Task 8
def show_palindrome(number,seperator = "") :
    str1 = ""
    for item in range(1,number+1) :
        str1 += str(item) + seperator
    for item in range(number-1,0,-1) :
        str1 += str(item) +seperator
    return str1
def show_palindromic_triangle(number) :
    for item in range(1, number+1) :
        print(" "*(2*number-2*item),end="")
        print(show_palindrome(item, " "))
show_palindromic_triangle(int(input('Enter a number :')))
```

```python
# Task 13
def function_name(operator,operand1,operand2) :
    if operator == '+' :
        return operand1 + operand2
    elif operator == '-' :
        return operand1 - operand2
    elif operator == '*' :
        return operand1 * operand2
    elif operator == '/' :
        return operand1 / operand2
operator,operand1,operand2 = input("Enter your operator :"), float(input("Enter the first operand :")), float(input("Enter the second operand :"))
print(function_name(operator,operand1,operand2))
```

```python
# Task 14
def function_name(string, position):
    string1 = string[0]
    remove = " "
    for item in range(1, len(string)):
        if item % position == 0:
            remove = remove + string[item]
        else:
            string1 += string[item]
    string1 = string1 + remove
    return string1
print(function_name(input("Enter your text: "), int(input("Enter position: "))))
```

```python
# Task 15
def function_name(order_menu,location = "Dhanmondi") :
    items = {"rice": 105, "potato": 20, "chicken": 250, "beef": 510, "oil": 85}
    price = 0
    for value in order_menu :
        if value in items :
            price += items[value]
    if location != "Dhanmondi" and price != 0 :
        price += 70
    elif location == "Dhanmondi" and price != 0 :
        price += 30
    return price
order_items = []
for value in input("Enter your grocery items :").lower().split(",") :
    order_items.append(value.strip())
print(function_name(order_items))
```

```python
# Task 18
def division(starting,ending,first_divisor,second_divisor):
    sum=0
    for i in range(starting,ending):
        if i%first_divisor==0 and i%second_divisor==0:
            continue
        elif i%first_divisor==0:
            sum=sum+i
        elif i % second_divisor == 0:
            sum = sum + i
    return sum
starting=int(input())
ending=int(input())
first_divisor=int(input())
second_divisor=int(input())

print(division(starting,ending,first_divisor,second_divisor))
```
