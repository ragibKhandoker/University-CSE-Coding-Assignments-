```python
# Task 8
str1 = input('Enter a dictionary :')
str1 = str1[1:-1]
list1 = str1.split(", ")
dict1 = {}
counter1 = 0
for item in list1 :
    list2 = item.split(":")
    key1 = list2[0][1:-1]
    value1 = int(list2[1])
    dict1[key1] = value1
    counter1 += 1
counter2 = 0
for value in dict1.values() :
    counter2 += value
avg1 = int(counter2/counter1)
print("Average is", avg1 )
```

```python
# Task 10
dict1 = {'sci fi': 12, 'mystery': 15, 'horror': 8, 'mythology': 10, 'young_adult': 4, 'adventure':14}
key1 = None
value1 = None
for item1,item2 in dict1.items():
    if key1 == None and value1 == None :
        key1 = item1
        value1 = item2
    else :
        if item2 > value1 :
            key1 = item1
            value1 = item2
print("The highest selling book genre is '" + str(key1) + "' and the number of books sold are", int(value1))
```

```python
# Task 11
str1 = input('Enter a string :')
dict1 = {}
for char in str1 :
    if char.isalpha() :
        char = char.lower()
        if char in dict1 :
            dict1[char] += 1
        else :
            dict1[char] = 1
print(dict1)
```

```python
# Task 14
dict1 = {'a':59 , 'b':-82 , 'c':5 , 'd':-81 , 'e':53}
for i in dict1:
    j = 0
    k = 22
    while j < 5:
        if j % 2 == 0:
            k = dict1[i] + j - (8 + k % 6) / 3
            dict1[i] = dict1[i]+ int(k)
        else:
            k = dict1[i] + j - (6 - k % 8) * 3
            dict1[i] = dict1[i] - int(k)
        j += 1
    print(int(k))
    print(i + " -> " + str(dict1[i]))
```
