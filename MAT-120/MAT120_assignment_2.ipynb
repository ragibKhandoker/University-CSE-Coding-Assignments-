# Determining maxima-minima of a Polynomial within a given interval

You have to look for the imaginary number and then append the roots to the list. Afterwards, checks if the elements of the list is within the interval with the help of if-else statement.

Find Maxima and Minima of $$
f = x^4 - \left(\frac{2}{3}\right)x^3 - 2x^2 + 2x
$$
within the given interval [0,2]

```python
import sympy as sp
x,y,z = sp.symbols("x y z")
f = x**4 - (sp.Rational(2)/sp.Rational(3))*x**3 - 2*x**2 + 2*x
df = sp.diff(f,x)
check = sp.solve(df,x)
print(check)
# check1= sp.solveset(df, x, sp.Interval(0,3))
# print(check1)
df2 = sp.diff(df,x)
sol =[]
#I havenot check if there is any imaginary number. But you have to check first.
for i in check:
  if i>=0 and i <=3:
    sol.append(i)

for i in sol:
  maxmin= df2.subs(x,i)
  if maxmin < 0:
    max= i
    print("etar jonno maxima pabo", max)
  elif maxmin > 0:
    min=i
    print("etar jonno minima pabo", min)

maxy= f.subs(x,max)
miny=f.subs(x,min)
print(f"maxima is {maxy.evalf()} and minima is {miny.evalf()}")
```

Plotting graph with **Matplotlib Library**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**4 - (2/3)*x**3 - 2*x**2 + 2*x

# Generate x values
x = np.linspace(-1,  3, 100)

# Calculate corresponding y values
y = f(x)

# Plot the graph
fig, ax = plt.subplots()
ax.plot(x, y)
# ax.set_xlim(-5, 5)
ax.set_ylim(0, 1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = x**4 - (2/3)*x**3 - 2*x**2 + 2*x')
plt.grid(True)
plt.show()
```

# Determining maxima-minima of a Polynomial within a given interval using solveset() Function

```python
import sympy as sp
from sympy import *
# init_printing(pretty_print=True)

x,y,z=sp.symbols('x y z')

expr1=sp.sin(x)

diff1=sp.diff(expr1,x)

check1=sp.solveset(diff1,x)
display(check1)
```

```python
f = sp.sin(x)
df = sp.diff(f,x)
#The most important line for the interval
check= sp.solveset(df, x, Interval(sp.pi, 2*sp.pi))
display(check)
```
