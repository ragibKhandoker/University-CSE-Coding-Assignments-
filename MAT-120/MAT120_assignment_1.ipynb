**Determine the Maxima and Minima of given function**

f(x) = Sin(x)

```python
from sympy.interactive.printing import init_printing
import sympy as sp
x,y,z = sp.symbols("x y z")
init_printing(pretty_print=True)
f = sp.sin(x) #writing the given function
df = sp.diff(f,x) #1st Derivative f'(X)
print(df)
check = sp.solve(df,x) #f'(x)=0
print(len(check))
sp.plot(f)
print(check) #critical points
df2 = sp.diff(df,x) #2nd derivative f''(x)
minmax1= df2.subs(x,check[0])
minmax2= df2.subs(x,check[1])
if minmax1 < 0:
  max= check[0]
  min= check[1]
elif minmax2 > 0:
  max= check[1]
  min= check[0]
maxval=f.subs(x,max)
minval= f.subs(x,min)
print(f'maxima is {maxval.evalf()} and minima is {minval.evalf()}')
```

Maxima and Minima of  $$ f = x^4 - 8x^3 + 22x^2 - 24x + 5 $$


```python
from sympy.interactive.printing import init_printing
import sympy as sp
x,y,z = sp.symbols("x y z")
init_printing(pretty_print=True)
f = x**4-8*x**3+22**2-24*x+5
df = sp.diff(f,x)
print(df)
check = sp.solve(df,x)
print(len(check))
print(check)
print(check[0].evalf())
print(check[1].evalf())
print(check[2].evalf())
df2 = sp.diff(df,x)
for i in check:
  if (sp.im(i)==0): #Check if there is any imaginary number.will not consider imaginary values
    minmax1 = df2.subs(x,i).evalf()
    print(minmax1)
    if minmax1 < 0:
      max = i
    elif minmax1 > 0:
      min=i
maxval=f.subs(x,max)
minval= f.subs(x,min)
print(f'maxima is {maxval.evalf()} and minima is {minval.evalf()}') #use evalf() to convert to numaric value
```
