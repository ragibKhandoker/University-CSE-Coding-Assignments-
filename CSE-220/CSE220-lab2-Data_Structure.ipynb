```python
#part 2
arr = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]
#task-1
def mean(arr):

    n = len(arr)
    sum = 0
    for i in arr:
        sum += i
    x = sum / len(arr)
    return x

print(mean(arr))
```

```python
#task-2
import math
def standard_deviation(arr):
    n = len(arr)
    var = 0
    z = mean(arr)
    for i in arr:
        x = i - z
        y = x**2
        var += y
    k = var/(len(arr)-1)
    f = k**0.5
    return f
print(standard_deviation(arr))
```

```python
#task-3
def StanDard(arr):
    x1 = mean(arr)
    x2 = standard_deviation(arr)
    c = 0
    upper_limit = x1 + 1.5*x2
    lower_limit = x1 - 1.5*x2
    for i in arr:
        if i > upper_limit or i < lower_limit:
            c += 1
    new_arr = [None] * c
    d =0
    for j in arr:
        if j > upper_limit or j < lower_limit:
            new_arr[d] = j
            d+= 1
    return new_arr
print(StanDard(arr))
```
