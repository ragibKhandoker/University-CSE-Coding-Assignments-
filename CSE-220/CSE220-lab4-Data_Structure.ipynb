```python
#Assignment
class Patient:
    def __init__(self,id,name,age,bloodg,next,prev):
        self.id = id
        self.name = name
        self.age = age
        self.bloodg = bloodg
        self.next = next
        self.prev = prev
class WRM:
    def __init__(self):
        self.head = Patient(None,None,None,None,None,None)
        self.head.next = self.head
        self.head.prev = self.head
    def RegisterPatient(self,id,name,age,bloodg): #creating a list
        last = self.head.prev
        new_node = Patient(id,name,age,bloodg,None,None)
        new_node.next = self.head
        new_node.prev = last
        self.head.prev = new_node
        last.next = new_node
        # temp = self.head
        # c = 0
        # while temp != self.head:
        #     print(c)
        #     c += 1
    def nodeAt(self,head,index):# nodeAt function
        temp = self.head.next
        c= 0
        while temp != self.head:
            if c == index:
                return temp
            temp = temp.next
            c += 1
        return None
    def ServePatient(self): # removing the 1st node patient
        index = 0
        remove_node = self.nodeAt(self.head,index)
        print(remove_node.name)
        previous_node = remove_node.prev
        next_node = remove_node.next
        previous_node.next = next_node
        next_node.prev = previous_node
        remove_node.next = None
        remove_node.prev = None
        return f"Patient being served"
        # return remove_node.id,remove_node.name,remove_node.age,remove_node.bloodg
    def ShowAllPatient(self): #count function
        temp = self.head.next
        #c = 0
        while temp != self.head:
            print(temp.name)
            temp = temp.next
            #c+= 1
        return f"All patient in sequence to serve"
    def CanDoctorGoHome(self):
        if self.head.next == self.head:
            return f"Yes,Doctor can go home"
        else:
            return f"No,Doctor cannot go home"
    def CancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head
        return f"Success"
obj = WRM()
while True:
    print("<-----Waiting Room Management----->")
    print("No.1- Add patient---")
    print("No.2- Serve Patient---")
    print("No.3- Show All patients---")
    print("No.4- Can Doctor Go Home---")
    print("No.5- Cancel All appointments---")
    print("No.6- Exit")

    option = int(input(" Please enter a number: "))
    if 1 <= option <= 6:
        if option == 6:
            break
        elif option == 1:
            id = int(input(" please enter patient id: "))
            name = str(input(" please enter patient name: "))
            age = int(input(" please enter patient age: "))
            bloodg = str(input(" please enter patient blood group: "))
            obj.RegisterPatient(id,name,age,bloodg)
            print("Successfully patient added")
        elif option == 2:
            print(obj.ServePatient())
        elif option == 3:
            print(obj.ShowAllPatient())
        elif option == 4:
            print(obj.CanDoctorGoHome())
        elif option == 5:
            print(obj.CancelAll())
    else:
        print("Invalid option. Please try again.")
print("Existing the program.....")
```
