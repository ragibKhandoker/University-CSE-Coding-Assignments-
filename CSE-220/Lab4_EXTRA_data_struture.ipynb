```python
#Assignment Part 1
class Patient:#node class
  #write a constructor
    def __init__(self,id,name,age,bloodgroup,prev = None,next = None):
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup
        self.prev = prev
        self.next = next
```

```python
class WRM:
#write a constructor
    def __init__(self):
        dh = Patient(None,None,None,None,None,None)
        self.head = dh
        self.head.prev = dh
        self.head.next = dh


    def registerPatient(self,id, name, age, bloodgroup):
        temp = self.head.prev
        new_patient = Patient(id,name,age,bloodgroup,next = self.head,prev = temp)
        temp.next = new_patient
        self.head.prev = new_patient

        #patient node class call for new patient appointment
        #temp=self.head




        return self.head

    def showAllPatient(self):
        flg = True
        temp = self.head.next
        while temp != self.head:
            print(f"Patient ID:{temp.id}")
            temp = temp.next


    def servePatient(self):#it means removing the first node
        temp = self.head.next
        if temp != self.head:
            self.head.next = temp.next
            temp.next.prev = self.head
            temp.next = None
            temp.prev = None
            print(temp.name)
        else:
            print('No patient to serve')


    def canDoctorGoHome(self):
        temp = self.head.next
        if temp == self.head:
            return True
        else:
            return False


    def cancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head
        print('Successfuly canceled')


    def ReverseTheLine(self):
        temp = self.head.next
        while temp != self.head:
            temp1 = temp.next
            temp.next = temp.prev
            temp.prev = temp1
            temp = temp1
        temp = self.head.next
        self.head.next = self.head.prev
        self.head.prev = temp
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    e.next = None
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```
