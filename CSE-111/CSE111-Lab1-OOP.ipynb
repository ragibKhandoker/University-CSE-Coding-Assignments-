```python
# Task 3
def UB_Jupmer(funcn):
    ref_list = [num for num in range(1, len(funcn))]
    abs_diff = []
    for i in range(len(funcn) - 1):
        elem = funcn[i] - funcn[i + 1]
        elem = abs(elem)
        abs_diff.append(elem)
    if sorted(ref_list) == sorted(abs_diff):
        return True
    else:
        return False
user_input = input('Enter your input :')
result = []
while user_input != "STOP":
    funcn = [int(num) for num in user_input.split()]
    if UB_Jupmer(funcn):
        result.append("UB Jumper")
    else:
        result.append("Not UB Jumper")
    user_input = input('Enter your input :')
for output in result:
    print(output)
```

```python
def key_map(string):
    key_map = {
        ".": 1, ",": 11, "?": 111, "!": 1111, ":": 11111,
        "A": 2, "B": 22, "C": 222,
        "D": 3, "E": 33, "F": 333,
        "G": 4, "H": 44, "I": 444,
        "J": 5, "K": 55, "L": 555,
        "M": 6, "N": 66, "O": 666,
        "P": 7, "Q": 77, "R": 777, "S": 7777,
        "T": 8, "U": 88, "V": 888,
        "W": 9, "X": 99, "Y": 999, "Z": 9999,
        " ": 0
    }
    return key_map.get(string, "(No-Key-For:{string})")
user_input = input().upper()
for char in user_input:
    key = key_map(char)
    print(key, end="")
```

```python
# Task 3
def invert(dictionary):
    dict1 = {}
    for item1, item2 in dictionary.items():
        if item2 in dict1:
            dict1[item2] = dict1[item2] + [item1]
        else:
            dict1[item2] = [item1]
    return dict1
str_input = input('Enter your input :')
dictionary = dict((x.strip(), y.strip())
    for x, y in (item.split(":") for item in str_input.split(", ")))
result = invert(dictionary)
print(result)
```

```python
# Task 4
def replace_domain(email, new, old = None) :
    newmail = ""
    status = ""
    if old != None :
        if old not in email :
            status = "Unchanged"
            newmail = email
        else :
            status = "Changed"
            for char in email :
                if char == "@" :
                    newmail += "@" + new
                    break
                newmail += char
    else :
        if new in email :
            status = "Unchanged"
            newmail = email
        else :
            status = "Changed"
            for char in email :
                if char == "@" :
                    newmail += "@" + new
                    break
                newmail += char
    return status + ": " + newmail
inp = input('Enter your input :')
email1, new1, old1 = tuple(str(elem) for elem in inp[2:-2].split(", "))
email = email1[:-1]
new = new1[1:-1]
old = old1[1:]
print( replace_domain(email, new, old))
```

```python
# Task 5
def palindrome(funcn) :
    var1 = ""
    for char in funcn :
        if char == " " :
            continue
        var1 += char
    if var1 == var1[::-1] :
        print("Palindrome")
    else :
        print("Not a palindrome")
funcn = input('Enter a string :')
funcn = funcn[1:-1]
palindrome(funcn)
```
