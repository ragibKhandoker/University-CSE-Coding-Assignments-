```python
# Task 9
class Batsman:
    def __init__(self, *args):
        self.name = "New Batsman"
        if len(args) == 2:
            runs, balls = args
        elif len(args) == 3:
            player, runs, balls = args
            self.name = player
        self.runs = runs
        self.balls = balls

    def setName(self, name):
        self.name = name

    def battingStrikeRate(self):
        return (self.runs) / (self.balls) * 100

    def printCareerStatistics(self):
        print(f"Name: {self.name}")
        print(f"Runs Scored: {self.runs}, Balls Faced: {self.balls}")

b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
# Task 13
class StudentDatabase:
    def __init__(self,name,id):
        self.name=name
        self.id=id
        self.grades={}
    def calculateGPA(self, sub, semester):
        total=0
        subjects=[]
        for item1 in sub:
            subname, subGrade = item1.split(':')
            total += float(subGrade)
            subjects.append(subname)
        gpa = round(total/len(sub),2)
        self.grades[semester]={tuple(subjects): gpa}

    def printDetails(self):
        print(f"Name: {self.name}\nID: {self.id}")
        for semester in self.grades:
            print(f"Courses taken in {semester}: ")
            for item2 in self.grades[semester].items():
                for item3 in item2[0]:
                    print(item3)
                print(f"Gpa: {item2[1]}")

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python
# Task 22
class Hotel :
    def __init__(self, place) :
        self.place = place
        self.age = 0
        self.phone = ''
        self.staff = 0
        self.guest = 0
        self.l1 = []
        self.l2 = []

    def addStuff(self, name1, age1, phone1 = '000') :
        self.name = name1
        self.age = age1
        self.phone = phone1
        self.l1_staff = [self.name, self.age, self.phone]
        self.l1.append(self.l1_staff)
        if self.name != None :
            self.staff += 1
            print(f"Staff with ID {self.staff} is added")

    def addGuest(self, name2, age2, phone2) :
        self.name = name2
        self.age = age2
        self.phone = phone2
        self.guest += 1
        self.l2_guest = [self.name, self.age, self.phone]
        self.l2.append(self.l2_guest)
        print(f"Guest with ID {self.guest} is created")

    def getStuffById(self, id1) :
        self.id = id1
        return f"Staff ID: {self.id}\nName: {self.l1_staff[0]}\nAge: {self.l1_staff[1]}\nPhone no.: {self.l1_staff[2]}"

    def getGuestById(self, id2) :
        self.id = id2
        return f"Guest ID: {self.id}\nName: {self.l2_guest[0]}\nAge: {self.l2_guest[1]}\nPhone no.: {self.l2_guest[2]}"

    def allStaffs(self) :
        print("All Staffs:")
        print("Number of Staff:", self.staff)
        var1 = 1
        for item in self.l1:
            print(f"Staff ID: {var1} Name: {item[0]} Age: {item[1]} Phone no: {item[2]}")
            var1 += 1

    def allGuest(self) :
        print("All Guest:")
        print("Number of Guest:", self.guest)
        var2 = 1
        for item in self.l2:
            print(f"Guest ID: {var2} Name: {item[0]} Age: {item[1]} Phone no: {item[2]}")
            var2 += 1

h = Hotel("Lakeshore")
h.addStuff( "Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest("Carol",35,"123")
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, "431")
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()
```

```python
# Task 23
class Author:
  def __init__(self,name="NA"):
    self.name=name
    self.numofbooks=0
    self.book={}
  def addBook(self,bname,genre):
    if self.name!="NA":
      if genre in self.book.keys():
        self.book[genre].append(bname)
      else:
        self.book[genre]=[bname]
      self.numofbooks+=1
    else:
      print('A book can not be added without author name')
  def setName(self,name):
    self.name=name
  def printDetail(self):
    print(f"Number of book(s): {self.numofbooks}")
    print(f"Author Name: {self.name} ")
    for k in self.book:
        print(k,':',end=" ")
        print(', '.join(self.book[k]))

a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```

```python
# Task 25
class Vaccine:
    def __init__(self, vn, pl, day):
        self.vn=vn
        self.pl=pl
        self.day=day
class Person:
    def __init__(self,name,age,prof="General Citizen"):
        self.name=name
        self.age=age
        self.prof=prof
        self.vac=""
        self.fdose=False
        self.sdose=False
    def pushVaccine(self,vacN,dose="1st Dose"):
        if dose=="1st Dose":
            if self.prof=="Student" or self.age>=25:
                self.vac=vacN
                self.fdose==True
                print(f"1st dose done for {self.name}")
            else:
                print(f"Sorry {self.name}, Minimum age for taking vaccines is 25 years now.")
        else:
            if self.vac.vn!=vacN.vn:
                print(f"Sorry {self.name}, you cant take 2 different vaccines")
            elif self.vac.vn==vacN.vn:
                self.sdose=True
                print(f"2nd dose done for {self.name}")

    def showDetail(self):
        print(f"Name: {self.name} Age: {self.age} Type: {self.prof}\nVaccine name: {self.vac.vn}")
        if self.sdose==True:
            print("1st dose: Given\n2nd dose: Given")
        else:
            print(f"1st dose: Given\n2nd dose: Please come after {self.vac.day} days")

astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
